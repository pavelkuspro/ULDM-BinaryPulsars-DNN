From the $S$ value and given fiducial values, we can determinate the $\lambda$ value.

```python
# compute the lambda value when mass, r, upsilon and SNR are given

def threshold_lambda_spin0l(m, r, upsasc, snr):

    m_val = m.to(u.eV).value
    omegb_ev_val =omega_b_ev.value
    k = m_val / omegb_ev_val
    psi =  np.array(psi_i(omega_b, Tasc, TOA_sim_array))

    q_vec = residuum_s0l(1, x, m, rhoDM, r, omega_b_ev, psi, upsasc)
    q_len = np.sqrt( np.dot(q_vec,q_vec) )

    lamb = snr * noise.to(u.s) / q_len.to(u.s)

    return lamb
```

We choose a subset of mass points:

```python
# considered mass interval
expon = np.linspace(-22, -21, 10) 
mass_list_short = 10**expon * u.eV

print(mass_list_short)
```

We compute the lambda value for each mass: 

```python
# a choice of DM phase and amplitude
upsasc_ex, r_ex = np.pi/4, 1 / np.sqrt(2)

# here, SNR is such that accuracy is 99%, 50%, 5%
lamb_s0l_array_p99_A = []
for m in np.array(mass_list_short)*u.eV:
    lamb = np.array(threshold_lambda_spin0l(m, r_ex, upsasc_ex, snr_s0l_p99_A))
    lamb_s0l_array_p99_A.append(lamb)
lamb_s0l_array_p99_A = np.array(lamb_s0l_array_p99_A)

lamb_s0l_array_p50_A = []
for m in np.array(mass_list_short)*u.eV:
    lamb = np.array(threshold_lambda_spin0l(m, r_ex, upsasc_ex, snr_s0l_p50_A))
    lamb_s0l_array_p50_A.append(lamb)
lamb_s0l_array_p50_A = np.array(lamb_s0l_array_p50_A)

lamb_s0l_array_p05_A = []
for m in np.array(mass_list_short)*u.eV:
    lamb = np.array(threshold_lambda_spin0l(m, r_ex, upsasc_ex, snr_s0l_p05_A))
    lamb_s0l_array_p05_A.append(lamb)
lamb_s0l_array_p05_A = np.array(lamb_s0l_array_p05_A)
```
